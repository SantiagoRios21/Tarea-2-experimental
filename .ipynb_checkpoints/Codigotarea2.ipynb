```python
# Solo para omitir warnings, en general NO deberÃ­an usarlo
import warnings
warnings.filterwarnings("ignore")

#import pyfits #modulo para leer archivos fits
from astropy.io import fits
import matplotlib.pyplot as plt #modulo para graficar
import numpy as np #este modulo es para trabajar con matrices como en matlab
import scipy as sp
from scipy import ndimage
from scipy.optimize import curve_fit
from astropy.stats import sigma_clip

import pandas as pd 
```

```python
cubo = fits.open("southgal_fixbadc.fits") #abrir objeto cubo de datos
data = cubo[0].data #extraer matriz de datos
header= cubo[0].header #extraer el header del archivo fits
```

```python
# funcion para obtener los valores de los ejes l, b, v
def values(h,j):
    N=h['NAXIS'+str(j)];
    val=np.zeros(N)
    for i in range(0,N):
        val[i] = (i+1-float(h['CRPIX'+str(j)]))*float(h['CDELT'+str(j)]) + float(h['CRVAL'+str(j)])
    return val
#Estos seran los tres arreglos con los valores reales de los tres ejes del cubo
velocidad=values(header,1)
longitud=values(header,2)
latitud=values(header,3)
```

```python

def velocidad_min(l,latitud, vs):
    #recorre latitud
    for q in range(len(latitud)):
        T1=data[q][l][:]
        rms=np.sqrt(np.mean(T1**2))   #calcula rms
        for w in range(len(velocidad)):
            if data[q][l][w]>=5*rms:  #buscamos que no sea ruido
                vs[q]=velocidad[w]    #guardamos la primera v donde T mayor a 5rms
                break
    return None
R0=8.5 #kPc
vsol=220 #km/s

def arreglos_():
    vmin=np.zeros(385)
    bvmin=np.zeros(385)
    R=np.zeros(385)
    for i in range(len(longitud)):
        vs=np.zeros(33)
        velocidad_min(i,latitud, vs)
        v1=vs[0]
        b1=latitud[0]
        for j in range(len(latitud)-1):
            if vs[j+1]<v1:
                v1=vs[j+1]
                b1=latitud[j+1]
        vmin[i]=v1
        bvmin[i]=b1
        R[i]=np.abs(R0*sp.sin(longitud[i]*sp.pi/180.)) #R0*sin(l)
    return R,vmin,bvmin

R,vmin,bmin = arreglos_()
    
def velocidad_(R,vmin):
    vR=np.zeros(385)
    wr=np.zeros(385)
    for i in range(385):
        wr[i]= vmin[i]/(R0*sp.sin(longitud[i]*sp.pi/180.)*3.086e+16)+vsol/(R0*3.086e+16)
        vR[i]= wr[i]*R0*3.086e+16*np.abs(sp.sin(longitud[i]*sp.pi/180.))
    return vR,wr
```

```python
vR,wr = velocidad_(R,vmin)
```

```python
plt.plot(R,vR)
plt.grid()
plt.title("Velocidad de rotacion en funcion de R")
plt.xlabel("R [Kpc]")
plt.ylabel("Vtan [km/s]")
plt.savefig('velocidad de rotacion ')
plt.show()
```

```python
G=4.302e-6

def masapuntual(R,M0):
    M=M0
    v=sp.sqrt(G*M/R)
    return v
mpuntual,covmpuntual=curve_fit(masapuntual, R, vR)

plt.plot(R,vR,label='datos')
plt.plot(R,masapuntual(R,mpuntual[0]),'k',label='Masa puntual')
plt.xlabel('R [Kpc]')
plt.ylabel('Velocidad [km/s]') 
plt.title('Fiteo para el modelo de masa de la galaxia')
plt.grid()
plt.legend()
plt.savefig('masa puntual')
plt.show()
print(mpuntual)
```

```python
def discouni(R,s):
    M=(R**2)*(np.pi)*s
    v=sp.sqrt(G*M/R)
    return v
discuni,covdiscouni=curve_fit(discouni, R, vR)

plt.plot(R,vR,label='datos')
plt.plot(R,discouni(R,discuni[0]),'k',label='Disco Unitario')
plt.xlabel('R [Kpc]')
plt.ylabel('Velocidad [km/s]') 
plt.title('Fiteo para el modelo de masa de la galaxia')
plt.grid()
plt.legend()
plt.savefig('disco unitario')

plt.show()
print(discuni)
```

```python
def esferauni(R,p):
    M=(R**3)*(np.pi)*p*(4/3)
    v=sp.sqrt(G*M/R)
    return v
esfeuni,covesfeuni=curve_fit(esferauni, R, vR)

plt.plot(R,vR,label='datos')

plt.plot(R,esferauni(R,esfeuni[0]),'k',label='Esfera unitaria')
plt.xlabel('R [Kpc]')
plt.title('Fiteo para el modelo de masa de la galaxia')
plt.ylabel('Velocidad [km/s]') 
plt.grid()
plt.legend()
plt.savefig('esfera unitaria')
plt.show()
print(esfeuni)
```

```python
def mpunt_discouni(R,s,M0):
    M=(R**2)*(np.pi)*s+M0
    v=sp.sqrt(G*M/R)
    return v
mpunt_disc,covmpunt_disc=curve_fit(mpunt_discouni, R, vR,bounds=(0,np.inf))

plt.plot(R,vR,label='datos')
plt.plot(R,mpunt_discouni(R,mpunt_disc[0],mpunt_disc[1]),'k',label='Masa puntual + disco unitario')
plt.xlabel('R [Kpc]')
plt.ylabel('Velocidad [km/s]') 
plt.grid()
plt.legend()
plt.title('Fiteo para el modelo de masa de la galaxia')
plt.savefig('masa puntual disco')
plt.show()
print(mpunt_disc)
```

```python
def mpunt_esferauni(R,p,M0):
    M=(R**3)*(np.pi)*p*4/3+M0
    v=sp.sqrt(G*M/R)
    return v
mpunt_esfe,covmpunt_esfe=curve_fit(mpunt_esferauni, R, vR)

plt.plot(R,vR,label='datos')
plt.plot(R,mpunt_esferauni(R,mpunt_esfe[0],mpunt_esfe[1]),'k',label='Masa puntual + esfera unitaria')
plt.xlabel('R [Kpc]')
plt.ylabel('Velocidad [km/s]') 
plt.title('Fiteo para el modelo de masa de la galaxia')
plt.grid()
plt.legend()
plt.savefig('masa puntual esfera')
plt.show()
print(mpunt_esfe)
```
